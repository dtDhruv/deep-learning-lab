```python
def calc_yin(w: list, x: list, b: int):
    out = 0
    for wi, xi in zip(w, x):
        out += wi * xi
    return out + b
```

```python
def calc_y_pred(yin):
    if yin == 0:
        return 0
    elif yin >= 1:
        return 1
    elif yin <= -1:
        return -1
```

```python
def calc(x1, x2, t, b, w1, w2, wb, yin, y_pred):
    i = 0
    while(i < 4):
        print("i: ", i)
        yin.append(calc_yin([w1[i], w2[i]], [x1[i], x2[i]], wb[i]))
        y_pred.append(calc_y_pred(yin[i]))
        if t[i] == y_pred[i]:
            cw1, cw2, cw3 = 0,0,0
        else:
            cw1, cw2, cw3 = t[i] * x1[i], t[i] * x2[i], t[i] * b
        print("y:", yin[i], y_pred[i])
        w1.append(cw1 + w1[i])
        w2.append(cw2 + w2[i])
        wb.append(cw3 + wb[i])
    
        i += 1
        print("w: ", w1[i], w2[i], wb[i], "\n")
    
    return [w1[-1]], [w2[-1]], [wb[-1]]
        
    
```

```python
def train(epochs, t, x1, x2):
    w1, w2, wb = [0], [0], [0]
    b = 1

    for e in range(epochs):
        print(f"epoch {e + 1}")
        yin = []
        y_pred = []
        w1, w2, wb = calc(x1, x2, t, b, w1, w2, wb, yin, y_pred)
```

```python
print("AND Gate: \n")
x1 = [1, 1, -1, -1]
x2 = [1, -1, 1, -1]
t = [-1, 1, -1, -1]
train(epochs=2, t=t, x1=x1, x2=x2)
```

```python
print("OR Gate: \n")
x1 = [1, 1, 0, 0]
x2 = [1, 0, 1, 0]
t = [1, 1, 1, -1]
train(epochs=5, t=t, x1=x1, x2=x2)
```
